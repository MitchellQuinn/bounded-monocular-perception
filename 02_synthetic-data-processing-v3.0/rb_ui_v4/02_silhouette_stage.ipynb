# RB Pipeline v4 Silhouette Stage
Use the launcher below and set Stage to `silhouette`.

```python
from pathlib import Path
import sys

def find_pipeline_root(start: Path | None = None) -> Path:
    candidate = (start or Path.cwd()).resolve()
    if candidate.is_file():
        candidate = candidate.parent

    for current in (candidate, *candidate.parents):
        if (current / 'rb_pipeline_v4').exists() and (current / 'input-images').exists():
            return current

        nested = current / 'synthetic-data-processing-v2.0'
        if (nested / 'rb_pipeline_v4').exists() and (nested / 'input-images').exists():
            return nested

    raise RuntimeError('Could not locate synthetic-data-processing-v2.0 root.')

PIPELINE_ROOT = find_pipeline_root()
if str(PIPELINE_ROOT) not in sys.path:
    sys.path.insert(0, str(PIPELINE_ROOT))

from rb_pipeline_v4.widgets import display_pipeline_launcher_v4
launcher = display_pipeline_launcher_v4(PIPELINE_ROOT)
```
